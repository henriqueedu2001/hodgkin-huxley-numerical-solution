```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n_values = [30, 50, 100, 200, 300, 500, 
            1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 90000,
            10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000,
            ]

rk_dataframes = [pd.read_csv(f'imgs/rk_out/out_rk_{n}.csv') for n in n_values]
euler_dataframes = [pd.read_csv(f'imgs/implicit_euler_out/out_imp_euler_{n}.csv') for n in n_values]
```

```python
def generate_table(dataframes):
    table = []
    for i, df in enumerate(dataframes):
        t, V, m, h, n = df['0'], df['1'], df['2'], df['3'], df['4']
        n_size = n_values[i]
        
        V_last, m_last, h_last, n_last = V.iloc[-1], m.iloc[-1], h.iloc[-1], n.iloc[-1], 
        table.append([n_size, V_last, m_last, h_last, n_last])
    
    return table
```

```python
rk_convergence_table = pd.DataFrame(generate_table(rk_dataframes))
euler_convergence_table = pd.DataFrame(generate_table(euler_dataframes))

rk_convergence_table.to_csv('euler_conv_table.csv')
euler_convergence_table.to_csv('rk_conv_table.csv')
```

```python

values = rk_values

last = values[-1]

p_values = []

for i,x in enumerate(values[:-2]):
    diff_1 = values[i] - last
    diff_2 = values[i+1] - last
    r1 = np.divide(diff_1, diff_2)
    r1 = np.absolute(r1)
    r1 = np.log(r1)
    r2 = np.log(values[i+1][1]/values[i][1])
    
    p = (r1/r2)[2:]
    
    p_values.append(p)

P = pd.DataFrame(p_values)
P
```
