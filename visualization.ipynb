```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n_values = [20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]

dataframes = [pd.read_csv(f'imgs/implicit_euler_out/out_imp_euler_{n}.csv') for n in n_values]
dataframes
```

```python
def plot_figure(df, name):
    t, V, m, h, n = df['0'], df['1'], df['2'], df['3'], df['4']
    
    left=0.1
    bottom=0.1
    right=0.9
    top=0.9
    wspace=0.4
    hspace=0.4
    
    plt.figure()
    plt.subplot(4, 1, 1)
    plt.subplots_adjust(left=left, bottom=bottom, right=right, top=top, wspace=wspace, hspace=hspace)
    
    plt.ylabel('$V$ (mV)')
    plt.xlabel('Tempo (ms)')
    plt.plot(t, V, label='V', color = 'black', linewidth=0.8)
    plt.legend()
    plt.subplot(4, 1, 2)
    plt.subplots_adjust(left=left, bottom=bottom, right=right, top=top, wspace=wspace, hspace=hspace)
    
    plt.plot(t, m, label='m', color = 'black', linewidth=0.8)
    plt.ylabel('$m$')
    plt.xlabel('Tempo (ms)')
    plt.legend()
    plt.subplot(4, 1, 3)
    plt.subplots_adjust(left=left, bottom=bottom, right=right, top=top, wspace=wspace, hspace=hspace)
    
    plt.plot(t, h, label='h', color = 'black', linewidth=0.8)
    plt.ylabel('$h$')
    plt.xlabel('Tempo (ms)')
    plt.legend()
    plt.subplot(4, 1, 4)
    plt.subplots_adjust(left=left, bottom=bottom, right=right, top=top, wspace=wspace, hspace=hspace)
    
    plt.legend()
    plt.plot(t, n, label='n', color = 'black', linewidth=0.8)
    plt.ylabel('$n$')
    plt.xlabel('Tempo (ms)')
    plt.legend()
    plt.subplots_adjust(left=left, bottom=bottom, right=right, top=top, wspace=wspace, hspace=hspace)
    # plt.show()
    plt.savefig(name)
```

```python
for i, df in enumerate(dataframes):
    n = n_values[i]
    plot_figure(df, f'imgs/implicit_euler_out/out_imp_euler_{n}.png')
```

```python
t_0, T = 2, 5.5
line_styles = ['-', '--', '-.', ':']

for i, df in enumerate(dataframes):
    t, V, m, h, n = df['0'], df['1'], df['2'], df['3'], df['4']
    plt.plot(t, V, label=f'n={n_values[i]}', linestyle=line_styles[i], color='black')
    plt.xlim(t_0, T)

plt.xlabel('Tempo (ms)')  # Rótulo do eixo x
plt.ylabel('V (mV)')  # Rótulo do eixo y

plt.legend()  # Adicionando legenda
plt.savefig('evolucao')
```
